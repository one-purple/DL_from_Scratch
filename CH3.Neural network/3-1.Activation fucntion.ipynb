# Sigmoid function

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
a = np.ones(len(x))
plt.plot(x, y)
plt.plot(x, a, linestyle="--")
plt.ylim(-0.1, 1.1)
plt.show()
```

# ReLU

```python
def relu(x):
    return np.maximum(0, x)

x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.show()
```
